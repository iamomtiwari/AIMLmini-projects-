```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load your dataset (assuming it's in a CSV file)
# Replace 'your_dataset.csv' with the actual file path or name
df = pd.read_csv('/content/FOOD-DATA-GROUP1.csv')
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
# Select a specific row (e.g., first row) to get the nutritional values
# We will filter the numeric columns only
# Select only the numeric columns
numeric_df = df.select_dtypes(include=[int, float])

# Check the number of columns
print(f"Number of numeric columns: {numeric_df.shape[1]}")

# Now select the first row of numeric data (height for the bar chart)
height = numeric_df.iloc[0].values  # Get the values for the bar heights

# Get the corresponding tick labels for the selected numeric columns
tick_label = numeric_df.columns

# Ensure that 'left' matches the number of tick labels and heights
left = list(range(len(tick_label)))
```

label for bars

# Tick labels based on column names (excluding the 'food' column)

```python
# Check that the lengths of the arrays match
print(f"Length of left: {len(left)}")
print(f"Length of height: {len(height)}")
```

```python
#plotting a bar chart
# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(left, height, tick_label=tick_label, color=['red', 'green', 'blue', 'yellow'] * (len(left) // 4 + 1))
# Add labels and title
plt.xlabel('Nutritional Categories')
plt.ylabel('Values')
plt.title('Nutritional Analysis for a Specific Food Item')
# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the chart
plt.show()
```

#Histogram

```python
# Select the column you want to plot (replace 'Caloric Value' with your actual column name)
column_name = 'Caloric Value'  # Change to your column name
data = df[column_name]

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')  # Adjust bins and colors as needed

# Add labels and title
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')

# Display the histogram
plt.show()
```

#Scatter Plot

```python

# Select the columns to plot (replace with your actual column names)
x_column = 'Caloric Value'  # X-axis data
y_column = 'Protein'        # Y-axis data

# Extract the data
x_data = df[x_column]
y_data = df[y_column]

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, color='blue', alpha=0.5)  # Adjust color and alpha as needed

# Add labels and title
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f'Scatter Plot of {x_column} vs. {y_column}')

# Display the scatter plot
plt.show()
```

#Subplot

```python

# Select columns for plotting
x_column = 'Caloric Value'
y_column = 'Protein'
data_column = 'Carbohydrates'  # Example for a histogram

# Extract the data
x_data = df[x_column]
y_data = df[y_column]
hist_data = df[data_column]

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 6))  # 1 row, 2 columns

# Scatter Plot
axs[0].scatter(x_data, y_data, color='blue', alpha=0.5)
axs[0].set_xlabel(x_column)
axs[0].set_ylabel(y_column)
axs[0].set_title(f'Scatter Plot of {x_column} vs. {y_column}')

# Histogram
axs[1].hist(hist_data, bins=20, color='skyblue', edgecolor='black')
axs[1].set_xlabel(data_column)
axs[1].set_ylabel('Frequency')
axs[1].set_title(f'Histogram of {data_column}')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

#Seaborn library

```python
import seaborn as sns

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='food', y='Caloric Value', data=df, palette='viridis')

# Set title and labels
plt.title('Bar Plot of Food vs. Caloric Value')
plt.xlabel('Food')
plt.ylabel('Caloric Value')

# Rotate x labels for better readability
plt.xticks(rotation=90)

# Show the plot
sns.despine()
plt.show()
```

```python
# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='food', y='Caloric Value', data=df, palette='viridis')

# Set title and labels
plt.title('Bar Plot of Food vs. Caloric Value')
plt.xlabel('Food')
plt.ylabel('Caloric Value')

# Rotate x labels for better readability
#plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate labels by 45 degrees, align to the right, adjust font size
plt.xticks(rotation=90, ha='center', fontsize=10)

# Show the plot
plt.tight_layout()  # Adjusts the layout so everything fits well
sns.despine()
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Caloric Value', y='Protein', data=df, color='blue', alpha=0.5)

# Set title and labels
plt.title('Scatter Plot of Caloric Value vs. Protein')
plt.xlabel('Caloric Value')
plt.ylabel('Protein')

# Show the plot
sns.despine()
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Caloric Value'], bins=20, kde=False, color='skyblue')

# Set title and labels
plt.title('Histogram of Caloric Value')
plt.xlabel('Caloric Value')
plt.ylabel('Frequency')

# Show the plot
sns.despine()
plt.show()
```
